```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from caribbean_map import *
```

```python
fig = plt.figure(dpi=300, figsize=(fs[0], 3))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

a = pd.read_csv("/home/xbxu/hycom/GLBv0.08/beach_partic_GLB.csv", header=None).values[:,:-1]
b = pd.read_csv("/home/xbxu/hycom/GLBv0.08/beach_partic_REGU.csv", header=None).values[:,:-1]
c = pd.read_csv("/home/xbxu/hycom/GLBv0.08/beach_partic_REGR.csv", header=None).values[:,:-1]
xm = np.linspace(0, 12, a.shape[0])
ym = np.linspace(2010, 2020, a.shape[1])

ax1.plot(xm, np.mean(a, 1), c='red', label="GLB")
ax1.plot(xm, np.mean(b, 1), c='green', label="UNIFORM")
ax1.plot(xm, np.mean(c, 1), c='blue', label="REALISTIC")
ax1.legend(frameon=False, loc=4,fontsize=8)
ax1.set_xlabel("Time after release [month]")
ax1.set_ylabel("Number of beached particles [%]")
ax1.set_ylim([0, 100])

ax2.plot(ym, a[182,:], c='red', label="GLB")
ax2.plot(ym, b[182,:], c='green', label="UNIFORM")
ax2.plot(ym, c[182,:], c='blue', label="REALISTIC")
ax2.set_xlabel("Release time [year]")
ax2.set_ylim([0, 100])
fig.savefig(f'figures/beaching_stats_comparison.png', bbox_inches='tight', format='png', dpi=600)
```
